# news.duckdb 탐색

```python
import duckdb

con = duckdb.connect('data/news.duckdb', read_only=True)
```

```python
# 테이블 목록 확인
con.execute("SHOW TABLES").fetchdf()
```

```python
# 각 테이블 스키마 확인
tables = con.execute("SHOW TABLES").fetchall()
for (table,) in tables:
    print(f"\n=== {table} ===")
    display(con.execute(f"DESCRIBE {table}").fetchdf())
```

```python
# 각 테이블 행 수 확인
for (table,) in tables:
    count = con.execute(f"SELECT COUNT(*) FROM {table}").fetchone()[0]
    print(f"{table}: {count:,} rows")
```

```python
# 각 테이블 샘플 데이터 (최근 5건)
for (table,) in tables:
    print(f"\n=== {table} (sample) ===")
    display(con.execute(f"SELECT * FROM {table} LIMIT 5").fetchdf())
```
